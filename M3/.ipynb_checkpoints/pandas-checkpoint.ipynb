```python
import pandas as pd
```

```python
# Direccionario de productos
datos = {'nombre':['hamburguesas', 'hotdogs', 'alitas','papas a la francesa','platanos fritos'],
         'precios':[50,25,50,35,30],
         'vendidos':[5,4,5,8,2],
          'disponibilidad':['viernes','sabado','sabados',['viernes','sabados'],['viernes','sabados']]}
```

```python
df = pd.DataFrame(datos)

```

## Ejercicios
- ! Muestrame cual es el precio maximo de los productos
- ¿Qué producto se vende menos ?
- Media de ventas 
- ¿Cual es la moda de los productos vendidos?
- Desviancion estandar de los precios 


```python
#Ejercicio uno 
df[["nombre", "precios"]].max()
```

```python
#Ejercicio dos
df.groupby("precios").min()
```

```python
df.groupby("precios").max()
```

```python
df["vendidos"].min()
```

```python
df["vendidos"].mode() //que numero se repite más//
```

```python
df["precios"].std()  //Que tanto varia en precio de los productos con respeto a la media //
```

```python
df["vendidos"].mean() // media//
```

```python
print("Precio maximo \n",df.groupby("nombre").max())
print("Productos menos vendidos\n",df["vendidos"].min())
print("Media de los productos vendidos",df["vendidos"].mean())
print("Moda de los productos vendidos",df["vendidos"].mode())
print("Desviacion estandar",df["precios"].std())
```
